```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
URL= ['https://www.yelp.ca/biz/pai-northern-thai-kitchen-toronto-5?osq=Restaurants',
     'https://www.yelp.ca/biz/ozzys-burgers-vaughan-vaughan',
     'https://www.yelp.ca/biz/batl-axe-throwing-vaughan']
```

```python
df= pd.DataFrame(columns= ['Resturant_name', 'No_reviews', 'reviewer', 'reviewer_link', 'reviewer_fig', 'comment_text', 
                              'review_score', 'review_date', 'reviewer_location'])
```

```python
for url in URL:
    response= requests.get(url)
    if response.status_code == 200:
        soup= BeautifulSoup(markup=response.text, features="html.parser")
    else:
        print('Error in getting data!')
    
    # Getting data 
    Resturant_name= soup.find(name="h1" , class_="css-1se8maq").text
    No_reviews= soup.find(name='a', class_='css-19v1rkv').text.split()[0][1:]
    #
    review_content= soup.find_all(name='li' , class_='css-1q2nwpv')
    for review in review_content:
        #
        reviewer= review.find(name='a' , class_='css-19v1rkv').text
        #
        reviewer_link= review.find(name='a' , class_='css-19v1rkv')['href']
        reviewer_link= '/'.join(url.split(sep='/')[:3]) + reviewer_link
        #
        reviewer_fig= review.find(name='img' , class_='css-1pz4y59')['src']
        #
        comment_text= review.find(name='span' , class_='raw__09f24__T4Ezm').text
        #
        review_score= review.find(name='div', class_='five-stars__09f24__mBKym five-stars--regular__09f24__DgBNj css-1jq1ouh')['aria-label'].split()[0]
        #
        review_date= review.find(name='span', class_='css-chan6m').text
        #
        reviewer_location= review.find(name='span', class_= 'css-qgunke').text
        
        df= df.append({'Resturant_name':Resturant_name, 'No_reviews':No_reviews, 'reviewer':reviewer,
                       'reviewer_link':reviewer_link,'reviewer_fig':reviewer_fig, 'comment_text':comment_text,
                       'review_score':review_score, 'review_date':review_date,
                       'reviewer_location':reviewer_location}, ignore_index= True)
        df.to_csv('my_first_Data.csv')
        
```
